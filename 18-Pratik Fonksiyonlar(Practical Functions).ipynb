```python
def divideNumber(number):
    return number / 2

myList = [3,5,7,10,20,30]

myResultList = []
for num in myList:
    myResultList.append(divideNumber(num))
print(myResultList)
```

```python
#map fonksiyonu, bir fonksiyonu bir iterable'ın her elemanına uygulamak için kullanılır. 
# Sonuç olarak, map fonksiyonu bir map nesnesi döndürür, bu nedenle sonucu listeye dönüştürmek için list() fonksiyonunu kullanabiliriz.

help(map)
list(map(divideNumber,myList))

def controlString(string):
    return "Mustafa" in string

controlString("Mustafa ÇEÇEN")
controlString("ÇEÇEN")

myStringList = ["Mustafa", "Mustafa ÇEÇEN","ÇEÇEN","Mustafa"]

list(map(controlString,myStringList))
```

```python
#filter fonksiyonu, bir fonksiyonu bir iterable'ın her elemanına uygulayarak, fonksiyonun True döndürdüğü elemanları içeren bir filter nesnesi döndürür.
# Sonuç olarak, filter fonksiyonu bir filter nesnesi döndürür,
# bu nedenle sonucu listeye dönüştürmek için list() fonksiyonunu kullanabiliriz.

list(filter(controlString,myStringList))
```

```python
#lambda fonksiyonu, tek bir ifade içeren anonim bir fonksiyon oluşturmak için kullanılır.
# Lambda fonksiyonları, genellikle kısa ve basit işlemler için kullanılır ve gen

multiplyLambda = lambda num : num * 3

type(multiplyLambda)

multiplyLambda(20)

result = multiplyLambda(10)

numList = [10,20,30,40,50]

list(map(lambda num:num/4,numList))
```
