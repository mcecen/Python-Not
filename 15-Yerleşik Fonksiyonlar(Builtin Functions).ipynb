```python
my_list = [20,30,40,50,60]
for ix in range(len(my_list)):
    print(ix)
    print(my_list[ix])
```

```python
my_list = [20,30,40,50,60]

for element in enumerate(my_list):
    print(element)

for (ix,value) in enumerate(my_list):
    print(value)
```
